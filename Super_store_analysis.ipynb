```python
def unique_categories(data):
  for i in data.columns:
    if not any(isinstance(value,(int,float)) for value in data[i]):
      v=data[i].unique()
      print('\nUnique Categories of {} are:\n'.format(i))
      for j in v:
        print(f"{j}\n")
      plt.figure()
      data[i].value_counts().plot(kind='pie', autopct='%1.1f%%')
      plt.title(f"{i} Distribution")
      plt.ylabel("")
      plt.axis('equal')
      plt.show()
```

```python
data.drop('Country',axis=1,inplace=True)
unique_categories(data)
```

```python
df1=data.groupby('Region')['Sales'].mean()
df2=data.groupby('Ship Mode')['Sales'].mean()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
df1.plot(kind='bar',xlabel='Region',ylabel='Mean Sales', title="Average sales in each Region")
plt.subplot(1,2,2)
df2.plot(kind='bar',xlabel='Ship Mode',ylabel='Mean Sales',title="Average sales in each Ship Mode")
plt.subplots_adjust(wspace=0.5)
plt.show()
print(f'\nMaximum No.of sales are in {df1.idxmax()}')
```

```python
df3=data.groupby('Category')['Sales'].mean()
plt.figure()
df3.plot(kind='pie',autopct='%1.1f%%',title='Percentage of sales in each category',ylabel='')
plt.show()
print('Therefore Maximum No.of sales are in {} category'.format(df3.idxmax()))
```

```python
loss_data=data[data['Profit']<0]
loss_data=loss_data.copy()
loss_data.rename(columns={'Profit':'Loss'},inplace=True)
loss_data['Loss']=loss_data['Loss'].abs()
```

```python
loss_data
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
for i in loss_data.columns:
  if loss_data[i].dtype == 'object':
    loss_data[i]=encode.fit_transform(loss_data[i])
```

```python
loss_data
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(loss_data.corr(),annot=True,cmap='coolwarm',cbar=True)
# plt.title='Correlation Matrix'
plt.show()
```

```python
data_frame=loss_data[['Sub-Category','Sales','Quantity','Loss']]
```

```python
data_frame.head()
```

```python
data_frame=data_frame.copy()
data_frame['Sub-Category']=encode.inverse_transform(data_frame['Sub-Category'])
```

```python
data_frame.head()
```

```python
category_grouped_data=data_frame.groupby(['Sub-Category'])['Loss'].sum()
sales_grouped_data=data_frame.groupby(['Sub-Category'])['Sales'].sum()
```

```python
category_grouped_data
```

```python
sales_grouped_data
```

```python
# plt.figure()
# grouped_data.value_counts().plot(kind='pie',autopct='%1.1f%%')
# plt.ylabel("")
# plt.title('Loss occured for each product categories')
# plt.axis('equal')
# plt.show()


plt.figure(figsize=(8,8))
plt.pie(category_grouped_data,labels=category_grouped_data.index,autopct="%1.1f%%")
plt.title('Loss occured for each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_grouped_data,labels=sales_grouped_data.index,autopct="%1.1f%%")
plt.title('Sales in each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

**Profit Analysis**

```python
profit_data=data[data['Profit']>0]
profit_data=profit_data.copy()
profit_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
profit_encode=LabelEncoder()
for i in profit_data.columns:
  if profit_data[i].dtype == 'object':
    profit_data[i]=profit_encode.fit_transform(profit_data[i])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(profit_data.corr(),annot=True,cmap='coolwarm',cbar=True)
# plt.title='Correlation Matrix'
plt.show()
```

```python
profit_data_frame=profit_data[['Category','Sales','Quantity','Profit']]
```

```python
profit_data_frame = profit_data_frame.copy()
profit_data_frame['Category'] = profit_encode.inverse_transform(profit_data_frame['Category'])
```

```python
category_grouped_data_p=profit_data_frame.groupby(['Category'])['Profit'].sum()
sales_grouped_data_p=profit_data_frame.groupby(['Category'])['Sales'].sum()
quantity_grouped_data_p=profit_data_frame.groupby(['Category'])['Quantity'].sum()
```

```python
category_grouped_data_p
```

```python
sales_grouped_data_p
```

```python
quantity_grouped_data_p
```

```python
plt.figure(figsize=(8,8))
plt.pie(category_grouped_data_p,labels=category_grouped_data_p.index,autopct="%1.1f%%")
plt.title('Profit for each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_grouped_data_p,labels=sales_grouped_data_p.index,autopct="%1.1f%%")
plt.title('Sales occured in each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(quantity_grouped_data_p,labels=quantity_grouped_data_p.index,autopct="%1.1f%%")
plt.title('Quantity of each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```
