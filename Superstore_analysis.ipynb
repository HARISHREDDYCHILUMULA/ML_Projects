**Data Visualization**


```python
from tabulate import tabulate
def underline_text(*texts):
    return "\033[4m" + " ".join(map(str, texts)) + "\033[0m"
def unique_categories(data):
    for i in data.columns:
        if i not in ['City', 'State']:
            if not any(isinstance(value, (int, float)) for value in data[i]):
                v = data[i].unique()
                table_data = [(underline_text(j),) for j in v]
                table = tabulate(table_data, headers=[f"Unique Categories of {i}"], tablefmt="grid")
                print(f"\n{table}\n")
                plt.figure()
                data[i].value_counts().plot(kind='pie', autopct='%1.1f%%')
                plt.title(f"{i} Distribution")
                plt.ylabel("")
                plt.axis('equal')
                plt.show()
```

# **Loss Analysis**

**Data Manipulation**

```python
loss_data=data[data['Profit']<0]
loss_data=loss_data.copy()
loss_data.rename(columns={'Profit':'Loss'},inplace=True)
loss_data['Loss']=loss_data['Loss'].abs()
```

```python
loss_data
```

**Data Transformation**

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
for i in loss_data.columns:
  if loss_data[i].dtype == 'object':
    loss_data[i]=encode.fit_transform(loss_data[i])
```

```python
loss_data
```

**Correlation map**

```python
plt.figure(figsize=(10,8))
sns.heatmap(loss_data.corr(),annot=True,cmap='coolwarm',cbar=True)
plt.show()
```

**Data Pre-processing**

```python
data_frame=loss_data[['Sub-Category','Sales','Quantity','Loss']]
```

```python
data_frame.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(data_frame.corr(),annot=True,cmap='coolwarm',cbar=True)
plt.show()
```

```python
data_frame=data_frame.copy()
data_frame['Sub-Category']=encode.inverse_transform(data_frame['Sub-Category'])
```

```python
data_frame.head()
```

```python
category_grouped_data=data_frame.groupby(['Sub-Category'])['Loss'].sum()
sales_grouped_data=data_frame.groupby(['Sub-Category'])['Sales'].sum()
```

```python
category_grouped_data
```

```python
sales_grouped_data
```

**Data Visualization**

```python
plt.figure(figsize=(8,8))
plt.pie(category_grouped_data,labels=category_grouped_data.index,autopct="%1.1f%%")
plt.title('Loss occured for each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_grouped_data,labels=sales_grouped_data.index,autopct="%1.1f%%")
plt.title('Sales in each product categories')
plt.ylabel('')
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')
plt.show()
```

**Multi Criteria Decision Analysis**
1. Define Criteria Weights (make sure they add up to 1)
2. Normalize Criteria (Min-Max normalization)
3. Calculate the Overall Score
4. Rank the Categories

```python
sub_categories = [str(x) for x in data_frame.iloc[:,0]]
total_sales = [x for x in data_frame.iloc[:,1]]
quantity = [x for x in data_frame.iloc[:,2]]
loss = [x for x in data_frame.iloc[:,3]]

# Step 1:
weight_sales = 0.4
weight_quantity = 0.3
weight_loss = 0.3

# Step 2:
normalized_sales = [(x - min(total_sales)) / (max(total_sales) - min(total_sales)) for x in total_sales]
normalized_quantity = [(x - min(quantity)) / (max(quantity) - min(quantity)) for x in quantity]
normalized_loss = [(x - min(loss)) / (max(loss) - min(loss)) for x in loss]

# Step 3:
overall_scores = [weight_sales * s + weight_quantity * q + weight_loss * p for s, q, p in zip(normalized_sales, normalized_quantity, normalized_loss)]

# Step 4:
loss_ranked_categories = [x for _, x in sorted(zip(overall_scores, sub_categories), reverse=True)]

print("Defend your profits: The Product Category That May Decrease Your Financial Gains is", underline_text(loss_ranked_categories[0]),'!')
```

```python
ranked_data = pd.DataFrame({"Category": loss_ranked_categories})
unique_values = ranked_data["Category"].unique()
ranks_dict = {category: rank for rank, category in enumerate(unique_values, 1)}
ranked_data["Rank"] = ranked_data["Category"].map(ranks_dict)
unique_values_list = ranked_data.drop_duplicates(subset="Category").to_dict(orient="records")
print("\nUnique Categorical Ranks (Loss):")
print(tabulate(unique_values_list, headers="keys", tablefmt="grid", showindex=False))
```

**Findings on Losses:**
- We identified several product categories that are experiencing higher losses, such as Machines, Tables, and Binders.
- The analysis revealed that Sales, Quantity contribute significantly to losses in these categories.
- Recommendations: To mitigate losses, we suggest focusing on Demand forecasting, price optimization, optimizing inventory management, and exploring pricing strategies for the identified categories.

**Multi Criteria Decision Analysis**
1. Define Criteria Weights (make sure they add up to 1)
2. Normalize Criteria (Min-Max normalization)
3. Calculate the Overall Score
4. Rank the Categories

```python
categories = [x for x in profit_data_frame.iloc[:,0]]
total_sales = [x for x in profit_data_frame.iloc[:,1]]
quantity = [x for x in profit_data_frame.iloc[:,2]]
profit = [x for x in profit_data_frame.iloc[:,3]]

# Step 1:
weight_sales = 0.4
weight_quantity = 0.3
weight_profit = 0.3

# Step 2:
normalized_sales = [(x - min(total_sales)) / (max(total_sales) - min(total_sales)) for x in total_sales]
normalized_quantity = [(x - min(quantity)) / (max(quantity) - min(quantity)) for x in quantity]
normalized_profit = [(x - min(profit)) / (max(profit) - min(profit)) for x in profit]

# Step 3:
overall_scores = [weight_sales * s + weight_quantity * q + weight_profit * p for s, q, p in zip(normalized_sales, normalized_quantity, normalized_profit)]

# Step 4:
ranked_categories = [x for _, x in sorted(zip(overall_scores, categories), reverse=True)]

print("Profit Magnets: The Secret to Success Lies Within ",underline_text(ranked_categories[0],'Category!'))
```

```python
ranked_data = pd.DataFrame({"Category": ranked_categories})
unique_values = ranked_data["Category"].unique()
ranks_dict = {category: rank for rank, category in enumerate(unique_values, 1)}
ranked_data["Rank"] = ranked_data["Category"].map(ranks_dict)
unique_values_list = ranked_data.drop_duplicates(subset="Category").to_dict(orient="records")
print("\nUnique Categorical Ranks (Profit):")
print(tabulate(unique_values_list, headers="keys", tablefmt="grid", showindex=False))
```

**Findings on Profits:**
- Certain product categories have demonstrated exceptional profitability, including Art, Appliances.
- Key drivers of profits in these categories include product mix and margins, price optimization, operational efficiency, customer retention and loyalty and customer segmentation.
- Recommendations: To capitalize on high-profit categories, we propose further investing in product expansion, branding, competitor analysis, market expansion and considering cross-selling opportunities.
